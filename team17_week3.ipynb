```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
# read a diabetes_model.csv file from seaborn library
df = pd.read_csv('diabetes_model.csv')
# count plot outcomes
df
```

```python
sns.countplot(x='Outcome',data=df)
plt.show()
```

# Analyse effect of age on outcomes using boxplot

```python
sns.boxplot(x=df['Outcome'],y=df['Age'],data=df)
plt.title('Effect of age on outcome')
```

# Check for chances of getting diabetes if it is common in family,using diabetes pedigree function 

```python
sns.violinplot(x=df['Outcome'],y=df['DiabetesPedigreeFunction'],data=df)
```

# Check for the relationship with insulin & glucose and group the datapoints by outcomes

```python
sns.scatterplot(x=df['Insulin'],y=df['Glucose'],data=df,hue=df['Outcome'])
```

# Create a heatmap on Correlation of features in the data                                       

```python
sns.heatmap(df.corr())
plt.xticks(rotation=45)
```
